1. **What is Detectron2 and how does it differ from previous object detection frameworks?**

**Answer:**
Detectron2 is an open-source object detection and segmentation framework developed by Facebook AI Research (FAIR). It is built on PyTorch and is used for tasks like object detection, instance segmentation, and keypoint detection.

It differs from previous object detection frameworks in several ways:

* It is built on PyTorch, unlike older frameworks such as Detectron which used Caffe2, making it more flexible and user-friendly.
* It has a modular design, allowing easy customization of models and training processes.
* It provides improved performance with optimized implementations of models like Faster R-CNN and Mask R-CNN.
* It offers cleaner code, better documentation, and simpler configuration.
* It supports advanced features such as distributed training and mixed precision training.

Overall, Detectron2 is more efficient, flexible, and easier to use compared to earlier object detection frameworks.


2. **Explain the process and importance of data annotation when working with Detectron2.**

**Answer:**
Data annotation is the process of labeling images with relevant information such as object boundaries, classes, and keypoints so that a model can learn from the data.

**Process of Data Annotation:**

* **Data Collection:** Gather images relevant to the problem domain.
* **Labeling:** Annotate images using tools (e.g., bounding boxes, segmentation masks, or keypoints).
* **Format Conversion:** Convert annotations into formats supported by Detectron2, such as COCO JSON format.
* **Dataset Registration:** Register the annotated dataset in Detectron2 for training and evaluation.
* **Validation:** Check annotations for accuracy and consistency.

**Importance of Data Annotation:**

* **Model Accuracy:** High-quality annotations directly improve the performance of detection models.
* **Learning Object Features:** Proper labels help the model understand object shapes, positions, and classes.
* **Generalization:** Well-annotated diverse data helps the model perform better on unseen data.
* **Error Reduction:** Accurate annotations reduce false positives and false negatives.

4. **What are evaluation curves in Detectron2, and how are metrics like mAP and IoU interpreted?**

**Answer:**
Evaluation curves in Detectron2 are graphical representations used to assess the performance of object detection models. The most common curve is the **Precision-Recall (PR) curve**, which shows the trade-off between precision and recall at different confidence thresholds.

**Evaluation Curves:**

* **Precision-Recall Curve:** Plots precision (accuracy of positive predictions) against recall (ability to find all relevant objects).
* These curves help in understanding how well the model balances false positives and false negatives.

**Metrics Interpretation:**

* **IoU (Intersection over Union):**
  IoU measures the overlap between the predicted bounding box and the ground truth box.
  [
  IoU = \frac{\text{Area of Overlap}}{\text{Area of Union}}
  ]

  * Value ranges from 0 to 1.
  * Higher IoU indicates better localization accuracy.
  * A common threshold (e.g., 0.5) is used to determine correct detections.

* **mAP (mean Average Precision):**
  mAP is the average of Average Precision (AP) across all classes.

  * AP is calculated from the area under the Precision-Recall curve.
  * Higher mAP indicates better overall detection performance.
  * Detectron2 often reports mAP at different IoU thresholds (e.g., mAP@0.5, mAP@0.5:0.95).


5. **Compare Detectron2 and TFOD2 in terms of features, performance, and ease of use.**

**Answer:**

**Detectron2 vs TFOD2 (TensorFlow Object Detection API):**

* **Framework Base:**
  Detectron2 is built on PyTorch, while TFOD2 is based on TensorFlow. PyTorch offers more flexibility, whereas TensorFlow provides strong production support.

* **Features:**
  Detectron2 provides advanced implementations of state-of-the-art models like Faster R-CNN and Mask R-CNN with strong support for research experimentation.
  TFOD2 offers a wide range of pre-trained models, built-in pipelines, and integration with TensorFlow tools like TensorFlow Lite and TensorFlow Serving.

* **Performance:**
  Detectron2 generally delivers higher performance and faster training for research tasks due to optimized implementations.
  TFOD2 is highly optimized for deployment and scalability, especially in production environments.

* **Ease of Use:**
  Detectron2 has cleaner and more modular code, making it easier for researchers to customize models.
  TFOD2 can be more complex to set up due to its configuration files and dependencies but is well-documented for end-to-end pipelines.

* **Customization:**
  Detectron2 allows easy customization of architectures and training processes.
  TFOD2 supports customization but often requires more effort due to its structured pipeline.

* **Deployment:**
  Detectron2 is mainly research-oriented, with fewer built-in deployment tools.
  TFOD2 excels in deployment with support for mobile and cloud environments.

```python
# 7. Annotate a dataset using any tool of your choice and convert the
# annotations to COCO format for Detectron2.
import json
import os

def create_coco_format(image_dir, annotations_list):
    """
    annotations_list: list of dicts with {image_name, category_id, bbox [x,y,w,h]}
    """
    coco_data = {
        "images": [],
        "annotations": [],
        "categories": [
            {"id": 1, "name": "object_type_a"},
            {"id": 2, "name": "object_type_b"}
        ]
    }

    for i, ann in enumerate(annotations_list):
        # Add image info (normally you'd get width/height from the file)
        image_info = {
            "id": i,
            "file_name": ann['image_name'],
            "width": 800,
            "height": 600
        }
        if image_info not in coco_data["images"]:
            coco_data["images"].append(image_info)

        # Add annotation info
        annotation = {
            "id": i,
            "image_id": i,
            "category_id": ann['category_id'],
            "bbox": ann['bbox'], # [x, y, width, height]
            "area": ann['bbox'][2] * ann['bbox'][3],
            "iscrowd": 0
        }
        coco_data["annotations"].append(annotation)

    return coco_data

# Example usage:
sample_anns = [
    {"image_name": "img1.jpg", "category_id": 1, "bbox": [100, 150, 50, 50]},
    {"image_name": "img2.jpg", "category_id": 2, "bbox": [200, 50, 120, 80]}
]

coco_output = create_coco_format("images/", sample_anns)

with open('custom_annotations.json', 'w') as f:
    json.dump(coco_output, f, indent=4)

print("COCO annotations file 'custom_annotations.json' created successfully!")
with open('custom_annotations.json', 'r') as f:
    display(json.load(f))
```

```python
# 8. Write a script to download pretrained weights and configure paths for training in Detectron2.


from detectron2.config import get_cfg
from detectron2 import model_zoo
import os

# Initialize configuration
cfg = get_cfg()

# 1. Select a model architecture from the model zoo
# We use Faster R-CNN with a ResNet-50 backbone
cfg.merge_from_file(model_zoo.get_config_file("COCO-Detection/faster_rcnn_R_50_FPN_3x.yaml"))

# 2. Configure dataset names (these must be registered first)
cfg.DATASETS.TRAIN = ("my_dataset_train",)
cfg.DATASETS.TEST = ("my_dataset_val",)

# 3. Download and set pretrained weights
cfg.MODEL.WEIGHTS = model_zoo.get_checkpoint_url("COCO-Detection/faster_rcnn_R_50_FPN_3x.yaml")

# 4. Set solver (training) parameters
cfg.SOLVER.IMS_PER_BATCH = 2
cfg.SOLVER.BASE_LR = 0.00025
cfg.SOLVER.MAX_ITER = 300 # Low for demonstration
cfg.MODEL.ROI_HEADS.BATCH_SIZE_PER_IMAGE = 128
cfg.MODEL.ROI_HEADS.NUM_CLASSES = 2 # Based on our earlier example

# 5. Set output directory
os.makedirs(cfg.OUTPUT_DIR, exist_ok=True)
print(f"Configuration complete. Output directory: {cfg.OUTPUT_DIR}")
print(f"Pretrained weights path: {cfg.MODEL.WEIGHTS}")
```

```python
# 9. Show the steps and code to run inference using a trained Detectron2 model on a new image.
from detectron2.engine import DefaultPredictor
from detectron2.utils.visualizer import Visualizer
from detectron2.data import MetadataCatalog
import cv2
import os
from detectron2 import model_zoo
from google.colab.patches import cv2_imshow

# 1. Load weights and configure device
# Since CUDA is not available, we force the model to run on CPU
cfg.MODEL.DEVICE = 'cpu'

# Point to weights
cfg.MODEL.WEIGHTS = model_zoo.get_checkpoint_url("COCO-Detection/faster_rcnn_R_50_FPN_3x.yaml")
cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST = 0.5

# 2. Create predictor
predictor = DefaultPredictor(cfg)

# 3. Load a sample image
!wget -q http://images.cocodataset.org/val2017/000000039769.jpg -O input.jpg
im = cv2.imread("./input.jpg")

# 4. Run inference
outputs = predictor(im)

# 5. Visualize results
v = Visualizer(im[:, :, ::-1], MetadataCatalog.get(cfg.DATASETS.TRAIN[0]), scale=1.2)
out = v.draw_instance_predictions(outputs["instances"].to("cpu"))

# 6. Display
cv2_imshow(out.get_image()[:, :, ::-1])
print("Inference completed successfully on CPU!")
```

```python
import cv2
import os
import requests
import numpy as np
from detectron2.engine import DefaultPredictor
from detectron2.config import get_cfg
from detectron2 import model_zoo
from detectron2.utils.visualizer import Visualizer
from detectron2.data import MetadataCatalog
from google.colab.patches import cv2_imshow

# 1. Pipeline Configuration
cfg = get_cfg()
cfg.merge_from_file(model_zoo.get_config_file("COCO-Detection/faster_rcnn_R_50_FPN_3x.yaml"))
cfg.MODEL.DEVICE = "cpu"
cfg.MODEL.WEIGHTS = model_zoo.get_checkpoint_url("COCO-Detection/faster_rcnn_R_50_FPN_3x.yaml")
cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST = 0.5
cfg.MODEL.ROI_HEADS.NUM_CLASSES = 80

# Initialize Predictor
predictor = DefaultPredictor(cfg)

# 2. Download a Lion Image with User-Agent to avoid 403
img_url = "https://upload.wikimedia.org/wikipedia/commons/7/73/Lion_waiting_in_Namibia.jpg"
headers = {'User-Agent': 'Mozilla/5.0'}
response = requests.get(img_url, headers=headers)

if response.status_code == 200:
    nparr = np.frombuffer(response.content, np.uint8)
    image = cv2.imdecode(nparr, cv2.IMREAD_COLOR)

    if image is not None:
        # 3. Run Inference
        outputs = predictor(image)

        # 4. Visualize
        metadata = MetadataCatalog.get("coco_2017_val")
        v = Visualizer(image[:, :, ::-1], metadata=metadata, scale=0.8)
        out = v.draw_instance_predictions(outputs["instances"].to("cpu"))

        print("--- Wildlife Monitoring System: Lion Detection ---")
        cv2_imshow(out.get_image()[:, :, ::-1])

        classes = outputs["instances"].pred_classes.tolist()
        found_animals = [metadata.thing_classes[i] for i in classes]
        print(f"Detected Species: {found_animals}")
    else:
        print("Error: Failed to decode image.")
else:
    print(f"Error: Failed to download. Status code: {response.status_code}")
```
